# Generating noise

```python
%%time
h, w = 512, 512
N = 1000
noise = np.stack([get_rgb_fractal_noise(h, w) for _ in tqdm(range(N))])
```

```python
plt.imshow(np.moveaxis(noise[0], 0, -1));
```

```python
np.save('../data/train_noise.npy', noise[:900])
```

```python
np.save('../data/val_noise.npy', noise[900:])
```
