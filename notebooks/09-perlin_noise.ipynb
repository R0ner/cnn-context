# Perlin noise

```python
noise = np.stack(
    [generate_fractal_noise_2d((256, 256), (128, 128), 1, 0.5) for i in range(3)]
)
plt.imshow(np.moveaxis((noise - noise.min()) / (noise.max() - noise.min()), 0, -1));
```

```python
h, w = 224, 224

ph, pw = int(np.ceil(np.log2(h)).item()), int(np.ceil(np.log2(w)).item())

ph_res, pw_res = random.randint(0, ph - 1), random.randint(0, pw - 1)

octaves = random.randint(1, min(ph - ph_res, pw - pw_res))

h_res, w_res = 2 ** ph_res, 2 ** pw_res

h_gen, w_gen = 2 ** ph, 2 ** pw
noise = np.stack(
    [generate_fractal_noise_2d((h_gen, w_gen), (h_res, w_res), octaves, 0.5) for i in range(3)]
)[:, :h, :w]
plt.imshow(np.moveaxis((noise - noise.min()) / (noise.max() - noise.min()), 0, -1));
```
